*   Are all launch sites in proximity to the Equator line?
   
   * A- Yes, the launch sites are in very close proximity to coasts. Launch sites are often located near the coast for several important reasons
     * 1- Flight Path Over Water: Launching rockets involves inherent risks, including the possibility of failure during ascent. By positioning                launch         sites near the coast, rockets can be directed over the ocean, minimizing the risk to populated areas on land.
     * 2- Stage Separation: Many rockets have multiple stages that separate during ascent. The spent stages typically fall back to Earth. Coastal
          locations allow these stages to drop into the ocean rather than on land, reducing the risk of damage or injury.
   * B- Access to Large Open Areas
     * 1-Unobstructed Launch Path: Coastal areas offer wide, open spaces with fewer obstacles like tall buildings or mountains. This is critical for          ensuring a clear path during launch.
     * 2-Emergency Situations: In case of an emergency or abort scenario, having the ocean nearby provides a safer area for an emergency landing or           splashdown.
   * C- Orbital Mechanics:
     * 1- Proximity to the Equator: Many coastal launch sites are located closer to the equator to take advantage of the Earth's rotational speed.             The Earth's rotation provides an additional boost to the rocket's velocity, making it easier and more fuel-efficient to reach orbit.
     * 2- Flexibility in Launch Trajectories: Coastal locations provide more flexibility in choosing launch trajectories, allowing rockets to reach            a variety of orbits (polar, geostationary, etc.) without overflying populated areas.
   * D- Logistical Considerations:
     * 1- Transportation of Large Components: Rockets and their components are often too large to be transported by road or rail. Coastal locations            allow for the use of ships and barges to transport these massive components directly to the launch site.
    
   * ##### Examples of Coastal Launch Sites:

* 1- Kennedy Space Center, Florida, USA: Located on the east coast, rockets launched from here can head east over the Atlantic Ocean.
* 2- Vandenberg Space Force Base, California, USA: Located on the west coast, this site is ideal for polar orbit launches, with rockets heading           south over the Pacific Ocean.
* 3- Guiana Space Centre, French Guiana: Positioned near the equator, rockets launched from here benefit from the Earth's rotational speed.

  ##### In summary, the proximity of launch sites to the coast is a strategic decision that balances safety, orbital mechanics, and logistical efficiency.



*   Are all launch sites in very close proximity to the coast?


    * No, not all launch sites are located near the Equator. While there are advantages to placing launch sites close to the Equator, there are many       launch sites located at higher latitudes as well. Here's a breakdown of why some launch sites are near the Equator and why others are located        further away:

      #### Advantages of Launch Sites Near the Equator:

      Increased Rotational Velocity: The Earth rotates faster at the Equator (about 1,670 km/h or 1,040 mph), providing rockets with an additional         velocity boost. This makes it easier and more fuel-efficient to reach equatorial orbits and geostationary orbits.
      Efficiency in Reaching Geostationary Orbit: Geostationary orbits are directly above the Equator, so launching from an equatorial site reduces        the energy required to reach this orbit.
      Global Accessibility: Equatorial launch sites can launch rockets to a wide range of orbital inclinations, making them versatile.

     #### Examples of Equatorial Launch Sites:

     Guiana Space Centre (Kourou, French Guiana): Located near the Equator, this site is ideal for launching to geostationary orbit.
     Satish Dhawan Space Centre (India): Also close to the Equator, this site is used for a variety of orbital launches.

     #### Launch Sites Far from the Equator: Many launch sites are located far from the Equator, often for reasons related to national security,          scientific research, or specific orbital needs:

     ##### 1. Polar Orbits:

    Vandenberg Space Force Base (California, USA): Located at 34.7°N, Vandenberg is ideal for launching satellites into polar orbits, which require      a north-south trajectory. These orbits are used for Earth observation satellites that need to pass over the poles to scan the entire surface of      the Earth.
    Plesetsk Cosmodrome (Russia): Located at 62.9°N, this site is used primarily for launching satellites into polar orbits.
    ##### 2. National Security:

    Baikonur Cosmodrome (Kazakhstan): Located at 45.6°N, Baikonur is the world's first and largest operational space launch facility. It was chosen      for strategic reasons during the Soviet era and continues to be a key site for Russian and international launches.

    
    ##### 3. Scientific Research:

    Esrange Space Center (Sweden): Located at 67.9°N, this site is used for launching research rockets and balloons. Its high latitude makes it          ideal for studying the auroras and other phenomena in the polar regions.
    Tanegashima Space Center (Japan): Located at 30.4°N, this site is Japan's primary spaceport. Its location was chosen to maximize safety and          accessibility rather than proximity to the Equator.
   #### Conclusion: While launch sites near the Equator offer specific advantages, particularly for launching satellites into equatorial orbits,         many important launch sites are located at higher latitudes to accommodate different mission profiles, such as polar orbits, national security       considerations, and scientific research. The location of a launch site depends on the specific needs of the missions it serves.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
def add_distance_marker(site_map, origin, coordinates):
    marker_cluster = MarkerCluster()
    site_map.add_child(marker_cluster)

    for coordinate in coordinates:
        distance = calculate_distance(*origin, *coordinate)

        distance_marker = folium.Marker(
            coordinate,
            icon=DivIcon(
                icon_size=(20,20),
                icon_anchor=(0,0),
                html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
            )
        )
        marker_cluster.add_child(distance_marker)

def add_lines(site_map, origin, coordinates):
    for coordinate in coordinates:
            line=folium.PolyLine(locations=[origin, coordinate], weight=1).add_to(site_map)
        
```
